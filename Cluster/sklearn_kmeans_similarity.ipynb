```python
from sklearn.datasets import load_digits
from sklearn.cluster import KMeans
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

cluster_count=10

digits = load_digits()
features = digits.data
rownum, linenum = features.shape
column = np.array(['feature{}'.format(index + 1) for index in range(linenum)])
column
```

```python
# clsutering
model = KMeans(init="k-means++", n_clusters=cluster_count, random_state=0)
model.fit(features)
```

```python
# get labels, centers
labels = model.labels_
labels
centers = model.cluster_centers_
centers.shape
```

```python
# merge fetatures and labels
features_copy = np.column_stack( (np.copy(features) , labels))
print(features_copy.shape)
print(features_copy.dtype)
# merge centers and labels
centers_copy = np.column_stack( (np.copy(centers) , range(0, cluster_count)))
print(centers_copy.shape)
print(centers_copy.dtype)
# merge fetatures and centers
features_cluster = np.vstack(( 
    np.copy( np.insert( features_copy, features_copy.shape[1], 1, axis=1 ) ) , 
    np.copy( np.insert( centers_copy, centers_copy.shape[1], 0, axis=1 ) ) ))
# add column with name 'prediction, type'
column_copy = np.concatenate((np.copy(column), ['prediction', 'type']))
column_copy
```

```python
P1 = pd.DataFrame(features_cluster, columns=column_copy)
P1.head()
```

```python
for i in range(cluster_count) :
    print(i, len(P1[P1.prediction == i]))
```

```python
from sklearn.metrics.pairwise import cosine_similarity
temp = []
for i in range(cluster_count) :
    df = P1[P1.prediction == i]
    center = df[df.type == 0]
    cs = cosine_similarity(center, df)
    df['weight'] = cs[0]
    temp.append(df)
P2 = pd.concat(temp)
```
